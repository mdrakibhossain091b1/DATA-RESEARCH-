```python
!pip install numpy pandas matplotlib seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
gym= pd.read_csv('/content/drive/MyDrive/Dataset/gym_membership.csv')
print("GYM DATASET:\n")
print(gym.head())
print("\nGYM DATASET Last 5 Rows:\n")
print(gym.tail())
print("\nData Type:\n", gym.dtypes)
print("\nData Shape:", gym.shape)
print("\nData Column:\n", gym.columns.to_list)
```

```python
# Task 1: Basic Matplotlib (3 plots)
# 1. Bar chart of gym membership types
# Use: gym['abonoment_type'].value_counts()
abonoment_type_counts= gym['abonoment_type'].value_counts()
plt.bar(abonoment_type_counts.index, abonoment_type_counts.values)
plt.title("Abonoment Type Count")
plt.tight_layout()
plt.show()
gender_counts= gym['gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.title("Gender Count")
plt.tight_layout()
plt.show()

# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])
plt.hist(gym['avg_time_in_gym'], bins=20)
plt.title("Avg Time in GYM")
plt.tight_layout()
plt.show()

plt.hist(gym['Age'], bins=20)
plt.title("Age in GYM")
plt.tight_layout()
plt.show()
# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])
plt.scatter(gym['Age'], gym['visit_per_week'])
plt.title("Age vs visits per week")
plt.xlabel("Age")
plt.ylabel("visits per week")
plt.tight_layout()
plt.show()
```

```python
# 1. Box plot: Gender vs Age
sns.boxplot(data=gym, x='gender', y='Age')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title("Gender vs Age")
plt.xlabel('Gender')
plt.ylabel('Age')
plt.tight_layout()
plt.show()
# 2. Count plot: Days per week
plt.figure(figsize=(50, 15))
sns.countplot(data=gym, x='days_per_week')
plt.title("Days per week")
plt.xlabel('Days per week')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/Dataset/Days per week.png', dpi=300, bbox_inches='tight')
plt.show()


# 3. Bar plot: Average visits by membership type
sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title("Average visits by membership")
plt.xlabel('Abonoment Type')
plt.ylabel('Visit Per week')
plt.tight_layout()
plt.show()
```
